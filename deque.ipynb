### Deque from Collections

```python
from collections import deque
d=deque()
c=0
while c<8:
    print("1. Adding elements to deque")
    print("2. Adding element to left of deque")
    print("3. Deleting last element from deque")
    print("4. Deleting first element from deque")
    print("5. Deleting the element in the deque")
    print("6. Displaying the deque")
    print("7. Finding the index of the element")
    print("8. Exit")
    c=int(input("Enter the operation number to be performed: "))
    if c==1:
        d.append(int(input("Enter the element: ")))
    elif c==2:
        d.appendleft(int(input("Enter the element to add from left: ")))
    elif c==3:
        if len(d)>=1:
            d.pop()
        else:
            print("deque is empty")
    elif c==4:
        if len(d)>=1:
            d.popleft()
        else:
            print("deque is empty")
    elif c==5:
        try:
            d.remove(int(input("Enter the element to remove: ")))
        except:
            print("Element not found")
    elif c==6:
        print(d)
    elif c==7:
        try:
            print(d.index(int(input("Enter the element to find the index: "))))
        except:
            print("Index not found for that element in the deque")
    else:
        break
```
